```python
import numpy as np
import pandas as pd
```

```python
# Markov story writer AI by Ninad Sumant
les_mis = open('scriptData.txt').read()

corpus = les_mis.split()
print('Word Count : {}'.format(len(corpus)))
```

```python
def make_pairs(text_corpus):
    for i in range(len(text_corpus)-1):
        yield (text_corpus[i], text_corpus[i+1])
```

```python
pairs = make_pairs(corpus)

word_dict = {}

for word_1, word_2 in pairs:
    if word_1 in word_dict.keys():
        word_dict[word_1].append(word_2)
    else:
        word_dict[word_1] = [word_2]
word_dict
```

```python
first_word = np.random.choice(corpus)

while first_word.islower():
    first_word = np.random.choice(corpus)

chain = [first_word]

n_words = 200

for i in range(n_words):
    chain.append(np.random.choice(word_dict[chain[-1]]))

gen_text = ' '.join(chain)
print('[ Markov Chain Generated Text ]\n\n{}'.format(gen_text))
```
